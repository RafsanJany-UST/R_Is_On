# Cell 8 – Basic Statistics with chickwts Dataset

```r
# built-in dataset
head(chickwts)

# quantiles
quantile(chickwts[, 1])
quantile(chickwts$weight, probs = 0.05)

# mean
mean(chickwts$weight)
mean(chickwts$weight, trim = 0.05)

# median
median(chickwts$weight)

# variance & standard deviation
var(chickwts$weight)
sd(chickwts$weight)
sqrt(var(chickwts$weight))   # same as sd
```

# Cell 9 – Range & IQR (Interquartile Range)

```r
# range (min, max)
range(chickwts$weight)

# max - min
max(chickwts$weight) - min(chickwts$weight)

# using diff()
diff(range(chickwts$weight))

# IQR
IQR(chickwts$weight)
```

# Cell 10 – Frequency Tables & Pie Charts

```r
# frequency table for feed type
table(chickwts$feed)

# add margins (totals)
addmargins(table(chickwts$feed))

# add margins with mean (not very meaningful here, but matches slide idea)
addmargins(table(chickwts$feed), FUN = mean)

# pie charts
par(mfrow = c(1, 2))

pie(table(chickwts$feed))

pie(
  table(chickwts$feed),
  col = rainbow(length(unique(chickwts$feed)))
)
```

# t.test

```r
BMI<-read.table(("C:\\Users\\RAZER\\Documents\\GitHub_2\\R_Is_On\\R_exercise_Data\\R_exercise\\bmi.txt"),
                col.names=c("height","weight","year","religion","gender","marriage"))
bmi<-BMI$weight/(BMI$height/100)^2  #bmi=kg/m2

qqnorm(bmi)
qqline(bmi,col="red")
shapiro.test(bmi)

t.test(bmi,mu=20.7, alter="less")  #(H1: mu< 20.7)
t.test(bmi,mu=20.7) #two-sided 


wilcox.test(bmi, mu = 20.7)
wilcox.test(bmi, mu = 20.7, alternative = "less")
```

```r

#Dataset: apple_hair.dat (http://users.stat.ufl.edu/~winner/data/apple_hair.dat)
hair.df<-read.table("C:\\Users\\RAZER\\Documents\\GitHub_2\\R_Is_On\\R_exercise_Data\\R_exercise\\apple_hair.txt", header=F)
head(hair.df)
hair.df<-read.table(("C:\\Users\\RAZER\\Documents\\GitHub_2\\R_Is_On\\R_exercise_Data\\R_exercise\\apple_hair.txt"),
                    col.names=c("trt","tot0","tot6","tot.inc","term0","term6","term.inc"))
head(hair.df)



shapiro.test(hair.df$tot.inc[hair.df$trt == 1])
shapiro.test(hair.df$tot.inc[hair.df$trt == 2])

attach(hair.df)
shapiro.test(tot.inc[trt==1])
shapiro.test(tot.inc[trt==2])

#var.test#########################
var.test(tot.inc~trt,data=hair.df)
t.test(tot.inc~trt, data=hair.df, var.equal=F, alter="less") #one sided (H1:mu1-mu2 <0)
t.test(tot.inc~trt, data=hair.df, var.equal=T)  #two-sided (#H1:mu1-mu2  not equal 0)
boxplot(tot.inc~trt, data=hair.df) #data visualization



wilcox.test(tot.inc ~ trt, data = hair.df)
wilcox.test(tot.inc ~ trt, data = hair.df, alternative = "less")
```
